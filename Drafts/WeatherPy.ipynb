### Analysis
* The weather gets warmer towards the equator (0 degrees latitude). Between +/- 20 degrees latitude and 0 degrees latitude there seems to be little change in the measured temperature, probably reflecting the fact that the location of the specific city (altitude, closeness to the sea...) has an influence on its temperature.
* The 'drop' in temperature is less clear in the southenr hemisphere, probably reflecting seasonal variations (currently there it is still summer). There are also significantly more cities between 20 and 80 degrees latitude than below -20 degrees latitude - this is mainly due to the distribution of land on the globe. 
* There is no strong relationship between latitude and cloudiness. Certain 'bands' of cloudiness (0, 20, 40, around 90) can be observed, but this may very well be related to the way the data is reported. 
* There is no strong relationship between latitude and wind speed. 'Outliers (high wind sppeds) seem to be more common at very high altitudes, but because the number of outliers is relatively small, it will require more analysis to see if this is a general trend or just due to specific weather conditions.

```python
# Load cities from cities.csv and put them into a dataframe

file_one = "cities.csv"

file_one_df = pd.read_csv(file_one, encoding="ISO-8859-1")

city_weather_df = file_one_df.drop(columns=["City_ID", "Country", "Date"])

city_weather_df.head()
```

#### Latitude vs. Temperature Plot

```python
%matplotlib inline

#Setting up the plot - called for all three required plots.
def set_up_plt(axis, y_label):
    plt.figure(num = None, figsize = [12,6])
    plt.grid(b=True, which = 'major', axis = 'both')
    plt.axis(axis)
    plt.xlabel("Latitude")
    plt.ylabel(y_label)
    
```

```python

y_label = 'Max. Daily Temperature'
axis = [-60, 90, 20, 120]
set_up_plt(axis, y_label)

plt.scatter(city_weather_df['Lat'], city_weather_df['Max Temp'])

plt.title("Max. Daily Temperature (C) vs. Latitude")

plt.savefig("Max_Temp_Lat.png")
```

#### Latitude vs. Humidity Plot

```python
y_label = "% Humidity"
axis = [-60, 80, 0, 100]
set_up_plt(axis, y_label)

plt.scatter(city_weather_df['Lat'], city_weather_df['Humidity'])

plt.title("% Humidity vs. Latitude (03/04/19)")

plt.savefig("Hum_Lat.png")
```

#### Latitude vs. Cloudiness Plot

```python
y_label = "% Cloud Cover"
axis = [-60, 80, 0, 100]
set_up_plt(axis, y_label)

plt.scatter(city_weather_df['Lat'], city_weather_df['Cloudiness'])

plt.title("% Cloudiness vs. Latitude (03/04/19)")

plt.savefig("Cloud_Lat.png")
```

#### Latitude vs. Wind Speed Plot

```python
y_label = "Wind Speed (m/s)"
axis = [-60, 80, 0, 35]
set_up_plt(axis, y_label)

plt.scatter(city_weather_df['Lat'], city_weather_df['Wind Speed'])

plt.title("Wind Speed (m/s) vs. Latitude (03/04/19)")

plt.savefig("WSpeed_Lat.png")
```
